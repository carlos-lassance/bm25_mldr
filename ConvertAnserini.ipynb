```python
from datasets import load_dataset
import os
import json
from tqdm import tqdm
import numpy as np
```

```python
for language in ["ar","de","en","es","fr","hi","it","ja","ko","pt","ru","th","zh"]:

    split = 'test'
    queryset = load_dataset('Shitao/MLDR', language, split=split, trust_remote_code=True)
    
    # load corpus
    corpus = load_dataset('Shitao/MLDR', f'corpus-{language}', split='corpus', trust_remote_code=True)
    
    default_path = "collections/MLDR"
    folder = os.path.join(default_path,language)
    file = os.path.join(folder,"corpus.jsonl")
    os.makedirs(folder,exist_ok=True)
    with open(file,"w") as f:
        for data in tqdm(corpus):
            local_dict = dict(id=data["docid"],contents=data["text"])
            f.write(json.dumps(local_dict)+"\n")
    
    default_path = "queries/MLDR"
    folder = os.path.join(default_path,language)
    os.makedirs(folder,exist_ok=True)
    qrels_file = os.path.join(folder,"test_qrels.tsv")
    query_file = os.path.join(folder,"test_queries.tsv")
    with open(qrels_file,"w") as lf:
        with open(query_file,"w") as uf:
            for data in tqdm(queryset):
                text = data["query"].replace("\n","")
                query_string = "{}\t{}\n".format(data["query_id"],text)
                if "\n" in text:
                    print(data["query_id"])
                uf.write(query_string)
                for results in data["positive_passages"]:
                    doc_id = results["docid"]
                    qrel_string = "{} 0 {} 1\n".format(data["query_id"],doc_id)
                    lf.write(qrel_string)
```
